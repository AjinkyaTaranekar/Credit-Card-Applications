```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
# adding data set
credit_card_apps = pd.read_csv('credit_approval_data_set\credit_card_data.csv')
```

```python
credit_card_apps.head()
```

```python
credit_card_apps.describe()
```

```python
credit_card_apps.info()
```

```python
# identifying missing variables 
credit_card_apps.tail(25)
# missing variables = '?'
```

```python
#substituting '?' with NaN
credit_card_apps = credit_card_apps.replace(['?'],np.NaN)
# Inspecting the missing values again
credit_card_apps.tail(25)
```

```python
# substituting NaN
credit_card_apps.fillna(credit_card_apps.mean(), inplace=True)
credit_card_apps.tail(25)
```

```python
for i in credit_card_apps.columns:
    if credit_card_apps[i].dtypes == 'object':
        credit_card_apps = credit_card_apps.fillna(credit_card_apps[i].value_counts().index[0])
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
credit_card_apps =pd.DataFrame(credit_card_apps)
for i in credit_card_apps.columns:
    if credit_card_apps[i].dtypes=='object':
        credit_card_apps[i]=pd.DataFrame(label.fit_transform(credit_card_apps[i]))
        
credit_card_apps.info()
credit_card_apps.head(10)
```

```python
credit_card_apps = credit_card_apps.drop(columns=['01', '00202'], axis=1)
credit_card_apps = credit_card_apps.values

X,y = credit_card_apps[:,0:13] , credit_card_apps[:,13]

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.33,random_state=42)
pd.DataFrame(credit_card_apps).head(10)
```

```python
logreg = LogisticRegression()
from sklearn.preprocessing import MinMaxScaler

# Instantiate MinMaxScaler and use it to rescale X_train and X_test
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.fit_transform(X_test)
logreg.fit(rescaledX_train,y_train)
```

```python
y_pred = logreg.predict(rescaledX_test)

print("Accuracy of logistic regression classifier: ",logreg.score(rescaledX_test,y_test))
print(confusion_matrix(y_test,y_pred))
```

```python
tol = [0.01,0.001,0.0001,0.00001,0.000001,0.0000001]
max_iter = [100,150,200]
param_grid = dict(tol=tol, max_iter=max_iter)
```

```python
grid_model = GridSearchCV(estimator=logreg, param_grid=param_grid, cv=10)

rescaledX = scaler.fit_transform(X)
grid_model_result = grid_model.fit(rescaledX, y)

best_score, best_params = grid_model_result.best_score_,grid_model_result.best_params_
print("Best: %f using %s" % (best_score, best_params))
```
